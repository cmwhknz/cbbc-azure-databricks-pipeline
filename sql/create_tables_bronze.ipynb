# Cbbc List Table

```python
%sql
CREATE TABLE IF NOT EXISTS bronze.cbbc_list (
    issuer STRING,
    code STRING,
    name STRING,
    bull_bear STRING,
    strike STRING,
    call_level STRING,
    call_spot_difference_pct STRING,
    expiry_date STRING,
    last_trading_day STRING,
    entitlement_ratio STRING,
    last_price STRING,
    issuer_bid STRING,
    issuer_ask STRING,
    issuer_bid_ask_spread STRING,
    change_dollar STRING,
    change_pct STRING,
    gearing STRING,
    premium_pct STRING,
    outstanding_qty_m STRING,
    outstanding_qty_pct STRING,
    outstanding_qty_day_change_m STRING,
    turnover_kdollar STRING,
    funding_cost_dollar STRING,
    funding_cost_pct STRING,
    underlying_code STRING,
    underlying_spot STRING,
    avg_bid_ask_spread_tick STRING,
    one_tick_spread_duration_pct STRING,
    liquidity STRING,
    listing_date STRING,
    quote STRING,
    trade_date DATE,
    processed_at TIMESTAMP
)
USING DELTA
```

# Cbbc List Ingestion Log Table

```python
%sql
CREATE TABLE IF NOT EXISTS bronze.cbbc_list_ingestion_log (
    file_name STRING,
    trade_date STRING,
    status STRING,
    error_message STRING,
    processed_at TIMESTAMP
)
USING DELTA;
```

# Cbbc Summary Table

```python
%sql
CREATE TABLE IF NOT EXISTS bronze.cbbc_summary (
    stock_short_name STRING,
    stock_code STRING,
    contracts_bought_count STRING,
    average_price_per_contract_bought STRING,
    contracts_sold_count STRING,
    average_price_per_contract_sold STRING,
    contracts_outstanding_count STRING,
    total_issue_size STRING,
    percent_issue_outstanding STRING,
    trading_currency STRING,
    meets_active_quote_criteria STRING,
    trade_date DATE,
    processed_at TIMESTAMP
)
USING DELTA;
```

# Cbbc Summary Ingestion Log

```python
%sql
CREATE TABLE IF NOT EXISTS bronze.cbbc_summary_ingestion_log (
    file_name STRING,
    trade_date STRING,
    issuer STRING,
    status STRING,
    error_message STRING,
    processed_at TIMESTAMP
)
USING DELTA;
```
