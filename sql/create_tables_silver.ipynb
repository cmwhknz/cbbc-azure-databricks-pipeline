# Cbbc Stake Curated Table

```python
%sql
CREATE TABLE IF NOT EXISTS silver.cbbc_stake_curated (
    trade_date DATE,
    cbbc_code STRING,
    bull_bear STRING,
    call_level DOUBLE,
    underlying_code STRING,
    contracts_outstanding_count DOUBLE,
    listing_date STRING,
    net_contracts_change DOUBLE,
    net_cash_flow DOUBLE
)
USING DELTA;
```
